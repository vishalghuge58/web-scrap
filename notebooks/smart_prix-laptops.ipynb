```python
from bs4 import BeautifulSoup
```

```python
with open('lapto.html','r',encoding='utf-8') as f:
    html=f.read()
```

```python
soup=BeautifulSoup(html,'lxml')
```

```python
a=soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
len(a)
```

```python
a[0]
```

```python
import pandas as pd
import numpy as np
```

```python
price=[]
name=[]
ram=[]
generation=[]
threds=[]
harddisk=[]
graphicscard=[]
size=[]
os=[]
warrenty=[]
```

```python
for i in a:
   
    try:
         x=i.find('ul',{'class':"sm-feat specs"}).find_all('li')
    except:
        continue
        
    try:
        
        price.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
        
    name.append(i.find('h2').text)
    
    
    print(x)
    
    try:
        generation.append(x[0].text)
    except:
        generation.append(np.nan)
    try:
        threds.append(x[1].text)
    except:
        threds.append(np.nan)
    try:
        harddisk.append(x[2].text)
    except:
        harddisk.append(np.nan)
        
    try:
        graphicscard.append(x[3].text)
    except:
        graphicscard.append(np.nan)
    try:
        size.append(x[4].text)
    except:
        size.append(np.nan)
        
    try:
        os.append(x[5].text)
    except:
        os.append(np.nan)
    try:
        warrenty.append(x[6].text)
    except:
        warrenty.append(np.nan)
```

```python
df=pd.DataFrame({
    
    "price":price,
    'name':name,
    'generation':generation,
    'threds':threds,
    'harddisk':harddisk,
    'graphicscard':graphicscard,
    'size':size,
    'os':os,
    'warrenty':warrenty
    
    
    
    
    
})
```
