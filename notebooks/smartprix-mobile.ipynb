```python
with open ('smartprix.html','r',encoding='utf-8') as f:
    html=f.read()
    
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,'lxml')
```

```python
# soup2=BeautifulSoup(html,'html.parser')
```

```python
container=soup.find_all('div','sm-product has-tag has-features has-actions')
```

```python
len(container)
```

```python
import pandas as pd
import numpy as np
```

```python
model=[]
price=[]
rating=[]
sim=[]
processor=[]
ram=[]
battery=[]
display=[]
camera=[]
memory_card=[]
vesion=[]
for i in container:
    try:
        model.append(i.find('h2').text)
    except:
        model.append(np.nan)

    try:
        
        price.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)

    try:
        
        rating.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
        
    except:
        rating.append(np.nan)
    
    x=i.find('ul',{'class':"sm-feat specs"}).find_all('li')
    try:
        
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:
        
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
        
    except:
        display.append(np.nan)

    try:
        camera.append(x[5].text)
        
    except:
        camera.append(np.nan)
    try:
        memory_card.append(x[6].text)
        
    except:
        memory_card.append(np.nan)

    try:
        vesion.append(x[7].text)
        
    except:
        vesion.append(np.nan)
```

```python
len(model)
```

```python
len(rating)
```

```python
df=pd.DataFrame({
    'model':model,
    'price':price,
    'rating':rating,
'processor':processor,
'ram':ram,
'battery':battery,
'display':display,
'camera':camera,
'memory_card':memory_card,'vesion':vesion}
        )
```
